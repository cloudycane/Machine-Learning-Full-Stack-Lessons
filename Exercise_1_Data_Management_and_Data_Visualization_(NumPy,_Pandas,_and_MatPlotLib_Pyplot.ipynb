```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Part 1: Create a series. It can be any series. Then transform the series into dataframes. Label the columns and indexes. Find the mean, variations, average, mode, and standard deviation. See the results of the information from the dataframes transposing it.

Part 2: Once we have the dataframe. We can plot it. The plotting needs to have graphics such as axis labels and pointers/markers.

```python
dataframe_a = {"Name": ["John", "Stacy", "Richard", "Joseph"], "Roles":["Human Resource", "Secretary", "Head of Staff", "Chief Employee"]}
dft = pd.DataFrame(dataframe_a)
print(dft)
```

```python
dft = pd.DataFrame(dataframe_a, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4"])
print(dft)

# The DataFrames need to have numeric values in order to find the mean, std, etc.

dataframe_b = {"Name": ["John", "Stacy", "Richard", "Joseph"], "Salary":[2000, 5000, 5000, 1000]}
dft2 = pd.DataFrame(dataframe_b)

dft = pd.merge(dft, dft2)

print(dft)
```

```python
mean = dft["Salary"].mean()
std = dft["Salary"].std()
mode = dft["Salary"].mode()
var = dft["Salary"].var()

print(f"Mean:{mean}, STD:{std}, Mode: {mode}, Variation: {var}")
```

```python
#X axis is the employees while Y axis is the salary

plt.plot(dft['Name'], dft['Salary'])
plt.title("Employees' Salary")
```
